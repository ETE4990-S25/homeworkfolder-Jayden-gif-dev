Using Matplotlib create the following graphs


````python
# Imports Here
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
````

````python
#Example datas
data = np.random.randint(1, 100, 50)
x = np.linspace(1, 10, 50)
y = np.sin(x) * 10 + np.random.randint(-5, 5, 50)
categories = ['Barbarian', 'Sorceress', 'Necromancer', 'Paladin', 'Amazon']
category_values = np.random.randint(10, 100, len(categories))
````

````python
# Histogram
plt.figure()
plt.hist(data, bins=10, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
````

````python
# Line Plot
plt.figure()
plt.plot(x, y, marker='o', linestyle='-', color='red', label='Battle Score')
plt.title('Line Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# Scatter Plot
plt.figure()
plt.scatter(x, y, color='green', label='Battle Data')
plt.title('Scatter Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# Pie Chart
plt.figure()
plt.pie(category_values, labels=categories, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.show()
````

````python
# Area Plot
plt.figure()
plt.fill_between(x, y, color='skyblue', alpha=0.5)
plt.title('Area Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True)
plt.show()
````

````python
# 3d Plot - Bar Plot
plt.figure()
plt.bar(x[:10], y[:10], color='purple', alpha=0.7)
plt.title('3D Bar Plot (2D View)')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
````

Using the following file https://www.kaggle.com/datasets/gregorut/videogamesales

Create and sort the data by Name, Year, and NA_Sales

Do not use pandas

````python
import numpy as np

dtype = [('Name', 'U100'), ('Year_of_Release', 'i4'), ('NA_Sales', 'f4')]

valid_rows = []
with open("vgsales.csv", "r", encoding="utf-8") as f:
    next(f)  # Skip the header row
    for line in f:
        parts = line.strip().split(",")

        if len(parts) >= 5:
            name = parts[0].strip()

            try:
                year = int(parts[1].strip())
            except ValueError:
                year = 0  
            
            try:
                na_sales = float(parts[4].strip())
            except ValueError:
                na_sales = 0.0  

            valid_rows.append((name, year, na_sales))

data = np.array(valid_rows, dtype=dtype)

sorted_by_name = np.sort(data, order="Name")  
sorted_by_year = np.sort(data, order="Year_of_Release")
sorted_by_sales = np.sort(data, order="NA_Sales")[::-1]

def print_section(title, data_array):
    print("\n" + "="*50)
    print(f" {title.upper()} ")
    print("="*50)
    for row in data_array:
        print(row)

print_section("Sorted by Name", sorted_by_name)
print_section("Sorted by Year", sorted_by_year)
print_section("Sorted by NA Sales (Descending)", sorted_by_sales)
````
