Using Matplotlib create the following graphs


````python
# Imports Here
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
````

````python
#Example datas
data = np.random.randint(1, 100, 50)
x = np.linspace(1, 10, 50)
y = np.sin(x) * 10 + np.random.randint(-5, 5, 50)
categories = ['Barbarian', 'Sorceress', 'Necromancer', 'Paladin', 'Amazon']
category_values = np.random.randint(10, 100, len(categories))
````

````python
# Histogram
plt.figure()
plt.hist(data, bins=10, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
````

````python
# Line Plot
plt.figure()
plt.plot(x, y, marker='o', linestyle='-', color='red', label='Battle Score')
plt.title('Line Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# Scatter Plot
plt.figure()
plt.scatter(x, y, color='green', label='Battle Data')
plt.title('Scatter Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# Pie Chart
plt.figure()
plt.pie(category_values, labels=categories, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.show()
````

Using the following file https://www.kaggle.com/datasets/gregorut/videogamesales

Create and sort the data by Name, Year, and NA_Sales

````python
import numpy as np

# I'm only tracking the game's name, release year, and NA sales. Simple and mildly useful.
dtype = [('Name', 'U100'), ('Year_of_Release', 'i4'), ('NA_Sales', 'f4')]

valid_rows = set()

# Cracking open the CSV file like it's full of secrets (its not).
with open("vgsales.csv", "r", encoding="utf-8") as f:
    next(f)  # Skipping the header  Ive read enough titles for one lifetime.
    for line in f:
        parts = line.strip().split(",")
        if len(parts) >= 5:
            name = parts[0].strip()
            try:
                year = int(parts[1].strip())
            except ValueError:
                year = 0  # If year is missing, Ill just pretend its ancient.
            try:
                na_sales = float(parts[4].strip())
            except ValueError:
                na_sales = 0.0  # No sales? Sounds about right.
            valid_rows.add((name, year, na_sales))

# Turning the chaos into a structured NumPy array. Feels like progress.
data = np.array(list(valid_rows), dtype=dtype)

# Sorting games by name (numbers first, because why not), year, and then NA sales (top sellers at the top).
sorted_by_name = np.array(sorted(data, key=lambda x: (not x[0][0].isdigit(), x[0])))
sorted_by_year = np.sort(data, order="Year_of_Release")
sorted_by_sales = np.sort(data, order="NA_Sales")[::-1]

def print_section(title, data_array):
    print(f"\n--- {title} ---")
    for row in data_array:
        name = row['Name']
        year = row['Year_of_Release']
        sales = row['NA_Sales']
        print(f"{name:<40} Year: {year:<5} NA Sales: {sales:.2f}")

# Time to show off my sorting skills.
print_section("Sorted by Name (Numbers First)", sorted_by_name)
print_section("Sorted by Year", sorted_by_year)
print_section("Sorted by NA Sales (Descending)", sorted_by_sales)
````
