# Farm Animal Management Program

## Introduction
This Python program helps **Old McDonald** categorize his farm animals using **object-oriented programming (OOP)**. It allows the user to:
- **Add animals** to the farm.
- **Remove animals** from the farm.
- **View all animals** currently on the farm.
- **Hear the animal sounds** and learn what each one produces (milk, wool, eggs, etc.).

The program uses **classes and inheritance** to organize different types of animals.

---

## How the Code is Structured

### **Defining the Base Animal Class**
We start with a base `Animal` class that contains:
- `name`: The name of the animal.
- `age`: The age of the animal.
- `make_sound()`: A generic method that each subclass will override.

```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

        return "Some generic animal sound"


````python
# Base class for all animals
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def make_sound(self):
        return "Some generic animal sound"

# Subclass for pets
class Pet(Animal):
    def __init__(self, name, age):
        super().__init__(name, age)
        self.is_friendly = True

# Subclass for farm animals
class FarmAnimal(Animal):
    def __init__(self, name, age, produces):
        super().__init__(name, age)
        self.produces = produces  # e.g., milk, wool, meat, eggs

# Specific animal classes
class Dog(Pet):
    def __init__(self, name, age):
        super().__init__(name, age)
        self.good_boy = True
    
    def make_sound(self):
        return "Woof!"

class Cat(Pet):
    def __init__(self, name, age):
        super().__init__(name, age)
    
    def make_sound(self):
        return "Meow!"

class Cow(FarmAnimal):
    def __init__(self, name, age):
        super().__init__(name, age, "Milk")
    
    def make_sound(self):
        return "Moo!"

class Pig(FarmAnimal):
    def __init__(self, name, age):
        super().__init__(name, age, "Meat")
    
    def make_sound(self):
        return "Oink!"

class Goat(FarmAnimal):
    def __init__(self, name, age):
        super().__init__(name, age, "Milk and Wool")
    
    def make_sound(self):
        return "Baa!"

class Chicken(FarmAnimal):
    def __init__(self, name, age):
        super().__init__(name, age, "Eggs")
    
    def make_sound(self):
        return "Cluck!"

class Sheep(FarmAnimal):
    def __init__(self, name, age):
        super().__init__(name, age, "Wool")
    
    def make_sound(self):
        return "Baa!"

class Duck(FarmAnimal):
    def __init__(self, name, age):
        super().__init__(name, age, "Eggs and Feathers")
    
    def make_sound(self):
        return "Quack!"

# Farm class to manage animals
class Farm:
    def __init__(self):
        self.animals = []

    def add_animal(self, animal):
        self.animals.append(animal)

    def remove_animal(self, name):
        self.animals = [animal for animal in self.animals if animal.name != name]

    def display_animals(self):
        for animal in self.animals:
            print(f"{animal.name} ({animal.__class__.__name__}), Age: {animal.age}, Sound: {animal.make_sound()}, Produces: {getattr(animal, 'produces', 'N/A')}")
````

````python
# Function to handle user interaction
def farm_simulation():
    farm = Farm()
    animal_classes = {
        "dog": Dog, "cat": Cat, "cow": Cow, "pig": Pig, 
        "goat": Goat, "chicken": Chicken, "sheep": Sheep, "duck": Duck
    }

    while True:
        print("\nOptions:")
        print("1. Add an animal")
        print("2. Remove an animal")
        print("3. Display animals")
        print("4. Exit")
        choice = input("Enter your choice: ").strip()

        if choice == "1":
            animal_type = input("Enter animal type (dog, cat, cow, pig, goat, chicken, sheep, duck): ").strip().lower()
            if animal_type in animal_classes:
                name = input("Enter the animal's name: ").strip()
                age = input("Enter the animal's age: ").strip()
                if age.isdigit():
                    age = int(age)
                    farm.add_animal(animal_classes[animal_type](name, age))
                else:
                    print("Invalid age. Please enter a number.")
            else:
                print("Invalid animal type. Please choose from the list.")

        elif choice == "2":
            name = input("Enter the name of the animal to remove: ").strip()
            farm.remove_animal(name)

        elif choice == "3":
            farm.display_animals()

        elif choice == "4":
            print("Exiting farm simulation. Have a great day!")
            break

        else:
            print("Invalid choice. Please select a valid option.")

farm_simulation()
````
